# Exploratory Data Analysis

```python
ax = sns.countplot(x = 'Gender', data = df,hue = 'Gender',palette=['maroon','pink'], legend = False)
plt.title("Number of Buyers(Male Vs Female)")
plt.xlabel("Gender")
plt.ylabel("Number of Buyers")
for bars in ax.containers:
    ax.bar_label(bars)
plt.savefig('buyers_gender.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
ax = sns.countplot(data = df, x="Marital_Status", hue="Gender",palette=['maroon','pink'], legend = False)
plt.title("Number of Buyers (Married vs Unmarried)")
plt.xlabel("Marital Status")
plt.ylabel("Number of Buyers")
for bars in ax.containers:
    ax.bar_label(bars)
plt.savefig('marital_status_gender.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
sales_gender = df.groupby(df['Gender'], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)
sales_gender
```

```python
fig = sns.countplot(data = df, x="Age Group", hue = 'Gender',palette=['maroon','pink'], legend = False)
plt.title("Number of Buyers according to Age Groups")
plt.xlabel("Age Group")
plt.ylabel("Number of Buyers")
for bars in fig.containers:
    fig.bar_label(bars)
plt.savefig('age_group_gender.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
sales_age = df.groupby(df['Age Group'], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)
plt.title("Total Sales Amount by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Total Amount")
sns.barplot(x = 'Age Group', y = 'Amount', data = sales_age,  color='maroon')
plt.savefig('sales_amount_age_group.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
sales_state = df.groupby(["State"], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,4)})
ax.set_title('Top 10 States by Number of Orders')
ax.set_xlabel('State')
ax.set_ylabel('Number of Orders')
sns.barplot(data = sales_state, x = 'State',y = 'Orders',  color='maroon')
plt.savefig('top_10_states_by_orders.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
sales_state = df.groupby(["State"], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,4)})
ax.set_title('Top 10 States by Total Amount')
ax.set_xlabel('State')
ax.set_ylabel('Total Amount')
sns.barplot(data = sales_state, x = 'State',y = 'Amount',  color='maroon')
plt.savefig('top_10_states_by_amount.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
ax = sns.countplot(x = 'Occupation', data = df,  color='maroon')
plt.title("Number of Buyers according to Occupation")
ax.set_xlabel('Occupation')
ax.set_ylabel('Number of Buyers')
sns.set(rc={'figure.figsize':(15,4)})
for bars in ax.containers:
    ax.bar_label(bars)
plt.savefig('buyers_by_occupation.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
sales_occupation = df.groupby(["Occupation"], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(15,4)})
ax.set_title('Total Amount by Occupation')
ax.set_xlabel('Occupation')
ax.set_ylabel('Total Amount')
sns.barplot(data = sales_occupation, x = 'Occupation',y = 'Amount',  color='maroon')
plt.savefig('total_amount_by_occupation.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
ax = sns.countplot(x = 'Product_Category', data = df,  color='maroon')
sns.set(rc={'figure.figsize':(25,12)})
ax.set_title('Number of Products by Category')
ax.set_xlabel('Product Category')
ax.set_ylabel('Number of Products')
plt.xticks(rotation=45, ha='right')
for bars in ax.containers:
    ax.bar_label(bars)
plt.savefig('products_by_category.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
sales_cate = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(8)
sns.set(rc={'figure.figsize':(20,10)})
ax.set_title('Top 8 Product Categories by Total Amount')
ax.set_xlabel('Product Category')
ax.set_ylabel('Total Amount')
sns.barplot(data = sales_cate, x = 'Product_Category',y = 'Amount',  color='maroon')
plt.savefig('top_8_product_categories.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
top_orders = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
ax.set_title('Top 6 Products by Number of Orders')
ax.set_xlabel('Product ID')
ax.set_ylabel('Number of Orders')
sns.barplot(data = top_orders, x = 'Product_ID',y = 'Orders',  color='maroon')
plt.savefig('top_6_products_by_orders.png', dpi=300, bbox_inches='tight')
plt.show()
```
